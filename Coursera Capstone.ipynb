# Code for second Submit

```python
geo_df = pd.read_csv('https://cocl.us/Geospatial_data', index_col=0)
geo_df
```

```python
Lat = []
Long = []
for i in df['Postal Code']:
    Lat.append(geo_df.loc[i,'Latitude'])
    Long.append(geo_df.loc[i,'Longitude'])
df['Latitude'] = Lat
df['Longitude'] = Long
df.loc[:,['Latitude', 'Longitude']]
```

# Code for third Submit
###### clustering
###### visualize result

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
for i in range(2,8):
    k_means = KMeans(init="k-means++", n_clusters=i, n_init=12)
    k_means.fit(df.loc[:,['Latitude', 'Longitude']])
    fig = plt.figure(i, figsize=(15, 10))
    plt.scatter(df['Latitude'], df['Longitude'], marker='.')
    k_means_labels = k_means.labels_
    k_means_cluster_centers = k_means.cluster_centers_
    colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))

# create a plot
    ax = fig.add_subplot(1, 1, 1)

# loop through the data and plot the datapoints and centroids.
# k will range from 0-3, which will match the number of clusters in the dataset.
    for k, col in zip(range(len([[4,4], [-2, -1], [2, -3], [1, 1]])), colors):

    # create a list of all datapoints, where the datapoitns that are 
    # in the cluster (ex. cluster 0) are labeled as true, else they are
    # labeled as false.
        my_members = (k_means_labels == k)
    
    # define the centroid, or cluster center.
        cluster_center = k_means_cluster_centers[k]
        
    # plot the datapoints with color col.
        ax.plot(df['Latitude'][my_members], df['Longitude'][my_members], 'w', markerfacecolor=col, marker='.')
    
    # plot the centroids with specified color, but with a darker outline
        ax.plot(cluster_center[0], cluster_center[1], 'o', markerfacecolor=col,  markeredgecolor='k', markersize=6)

# title of the plot
    ax.set_title('KMeans')

# remove x-axis ticks
    ax.set_xticks(())

# remove y-axis ticks
    ax.set_yticks(())

# show the plot
    plt.show()
```
